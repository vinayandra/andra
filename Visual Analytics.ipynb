```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
#import data
df = pd.read_csv("C:\\Users\\KATNABA1\\Desktop\\Visual Analytics.csv")
df.head()
```

```python
df.plot(kind= 'scatter', x = 'Balance', y = 'Age')
```

```python
counts = df['Job Classification'].value_counts() # finding the count for each unique category

counts
```

```python
counts = df['Region'].value_counts()# finding the count for each Region
counts
```

```python
fig = plt.figure(figsize=(8,8)) # Define plot area
ax = fig.gca() # defining axis
counts.plot.bar(ax = ax)# use the plot.bar method on the counts data frame
ax.set_title('No.of Job Classification  by Region') # Give the plot main title
ax.set_xlabel('Region') # set text for x axis
ax.set_ylabel('Job Classification') # set text for y axis
```

```python
fig = plt.figure(figsize=(8,8)) # Define plot area
ax = fig.gca() # defining axis
df['Age'].plot.hist(ax = ax) # use the plot.bar method on the counts data frame
ax.set_title('Histogram of Age') # Give the plot main title
ax.set_xlabel('Age') # set text for x axis
ax.set_ylabel('Balance') # set text for y axis
```

```python
fig = plt.figure(figsize=(8,8)) # Define plot area
ax = fig.gca() # defining axis
df[['Job Classification', 'Age']].boxplot(by = 'Job Classification',ax = ax) # use the plot.bar method on the counts data frame
ax.set_title('Boxplot of Job Classification by Gender') # Give the plot main title
ax.set_xlabel('Job Classification') # set text for x axis
ax.set_ylabel('Age') # set text for y axis
```

```python
import seaborn as sns

sns.kdeplot(df['Age']) # kernel density plot
```

```python

sns.kdeplot(df['Balance']) # kernel density plot
```

```python
fig = plt.figure(figsize=(8,8)) # Define plot area
ax = fig.gca() # defining axis
sns.kdeplot(df['Age'], ax = ax) #use the KDE method on the counts data frame
ax.set_title('KDE plot of Age') # Give the plot main title
ax.set_xlabel('Age') # set text for x axis
ax.set_ylabel('Balance') # set text for y axis
```

```python
fig = plt.figure(figsize=(8,8)) # Define plot area
ax = fig.gca() # defining axis
sns.kdeplot(df[['Age', 'Balance']], ax = ax, cmap='Blues_d') #use the KDE method on the counts data frame
ax.set_title('KDE plot of Age and Balance') # Give the plot main title
ax.set_xlabel('Age') # set text for x axis
ax.set_ylabel('Balance') # set text for y axis
```

```python
fig = plt.figure(figsize=(8,8)) # Define plot area
ax = fig.gca() # defining axis
sns.set_style("whitegrid")
sns.violinplot(x = 'Job Classification', y = 'Balance', data = df, ax = ax)  #use the violinplot method on the counts data frame
ax.set_title('Violin plot for ballance by Gender') # Give the plot main title
ax.set_xlabel('JOb Classification') # set text for x axis
ax.set_ylabel('Balance') # set text for y axis
```

```python
sns.lmplot(x = 'Region', y = 'Age',
          data = df,
          hue = "Job Classification",
          palette="Set2", fit_reg = False)
```
